```python
# Step 5: Models to train..

model_lr = LinearRegression()
model_rf = RandomForestRegressor(random_state=42)
model_xgb = XGBRegressor(random_state=42)

```

```python
# Step 6: Train models..
model_lr.fit(X_train, y_train)
model_rf.fit(X_train, y_train)
model_xgb.fit(X_train, y_train)
```

```python
# Step 7: Predictions..
pred_lr = model_lr.predict(X_test)
pred_rf = model_rf.predict(X_test)
pred_xgb = model_xgb.predict(X_test)
```

```python
# Step 8: Evaluation Function..
def evaluate_model(y_test, y_pred):
    r2 = r2_score(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    mae = mean_absolute_error(y_test, y_pred)
    return r2, rmse, mae
```

```python
# Step 9: Evaluate all models..
results = {
    "Linear Regression": evaluate_model(y_test, pred_lr),
    "Random Forest": evaluate_model(y_test, pred_rf),
    "XGBoost": evaluate_model(y_test, pred_xgb),
}
```

```python
 # Step 10: Display Results..
for model, metrics in results.items():
    print(f"\nModel: {model}")
    print(f"R2 Score: {metrics[0]:.4f}")
    print(f"RMSE: {metrics[1]:.4f}")
    print(f"MAE: {metrics[2]:.4f}")
```

```python
# Step 11: Select Best Model (Based on R2 Score)
best_model_name = max(results, key=lambda x: results[x][0])
print(f"\nBest Performing Model: {best_model_name}")


# Select best model object
best_model = (
    model_lr if best_model_name == "Linear Regression" else
    model_rf if best_model_name == "Random Forest" else
    model_xgb
)
```

```python
# Step 12: Save Best Model using Joblib:
joblib.dump(best_model, "sales_prediction_model.pkl")
print("\nBest model saved successfully as 'sales_prediction_model.pkl'")
```
